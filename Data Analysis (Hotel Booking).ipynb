```python
#convert the reservation_status_date to datetime format
hotel_booking['reservation_status_date'] = pd.to_datetime(hotel_booking['reservation_status_date'], format='%d/%m/%Y')
```

```python
# Strip whitespace from column names
hotel_booking.columns = hotel_booking.columns.str.strip()
```

```python
# Check if the columns exist in the DataFrame
columns_to_drop = ['company', 'agent']
for col in columns_to_drop:
    if col not in hotel_booking.columns:
        print(f"Column '{col}' not found in DataFrame.")
```

```python
#drop the agent and company column
hotel_booking.drop(['company', 'agent'], axis = 1, inplace = True)
hotel_booking.dropna(inplace = True)
```

```python
#check for missing value
hotel_booking.isnull().sum()
```

```python
hotel_booking['adr'].plot(kind = 'box')
```

```python
#remove outlier from adr column
hotel_booking = hotel_booking[hotel_booking['adr']<5000]
```

# Data analysis and visualization

```python
cancelled_perc = hotel_booking['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled', 'Canceled'], hotel_booking['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()
```

```python
#visualization based on hotel and the canceled rate
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = hotel_booking, palette = 'Blues')
legend_labels = ax1. get_legend_handles_labels()
ax1.legend(loc='upper right', bbox_to_anchor=(1, 1))
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('Hotel')
plt.ylabel('Number of reservation')
plt.show()
```

```python
#Count of resort hotel

resort_hotel = hotel_booking[hotel_booking['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
#Count of city hotel

city_hotel = hotel_booking[hotel_booking['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
#reservation status per month
hotel_booking['month'] = hotel_booking['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = hotel_booking, palette = 'bright')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status per month', size = 20)
plt.xlabel('Month')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
#average daily rate for each month
plt.figure(figsize=(15, 8))
plt.title('ADR per month', fontsize=30)

# Filter the DataFrame and group by month, then sum the 'adr' column
data_to_plot = hotel_booking[hotel_booking['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index()

# Plot the data using a barplot
sns.barplot(x='month', y='adr', data=data_to_plot)

plt.xlabel('Month', fontsize=20)
plt.ylabel('ADR', fontsize=20)
plt.xticks(rotation=45)
plt.show()
```

```python
#Top 10 Countries with reservation canceled
canceled_data = hotel_booking[hotel_booking['is_canceled'] == 1]
top_10_country = canceled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 Countries with reservation canceled')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```

```python
#count of market segment
hotel_booking['market_segment'].value_counts()
```

```python
#percentage count
hotel_booking['market_segment'].value_counts(normalize = True)
```

```python
#percentage count for cancelled booking
cancelled_data['market_segment'].value_counts(normalize = True)
```

```python
#average daily rate
cancelled_hotel_booking_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_hotel_booking_adr.reset_index(inplace = True)
cancelled_hotel_booking_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = hotel_booking[hotel_booking['is_canceled'] == 0]
not_cancelled_hotel_booking_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_hotel_booking_adr.reset_index(inplace = True)
not_cancelled_hotel_booking_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_hotel_booking_adr['reservation_status_date'], not_cancelled_hotel_booking_adr['adr'], label = 'not_cancelled')
plt.plot(cancelled_hotel_booking_adr['reservation_status_date'], cancelled_hotel_booking_adr['adr'], label = 'cancelled')
plt.legend()
```
